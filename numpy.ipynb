```python
t=np.arange(0,12,0.5)
y1=np.sin(t)
y2=np.cos(t)
plt.grid()
plt.plot(t, y1, 'y--')
plt.plot(t, y2, 'r--', marker="*")
plt.xlabel("Time")
plt.ylabel("Amp")
plt.title("SinCos")
plt.legend(["Sin","Cos"])
```

### 문제 subplot 을 이용해서 2행 1열 구조 2개씩 4가지 색상 다르게 스타일 다르게 마커 서로 다르게 범례표시 x,y 제목 

```python
x=np.arange(0,5,1)
y1=x+1
y2=x+2
y3=x+3
y4=x+4

plt.subplot(1,2,1)
plt.grid()
plt.plot(x,y1, 'y--', marker="o")
plt.plot(x,y2, 'r-', marker="*")
plt.ylabel('Time')
plt.xlabel('Amp')

plt.subplot(1,2,2)
plt.grid()
plt.plot(x,y3, 'g-', marker="v")
plt.plot(x,y4, 'b:', marker=".")
plt.ylabel('Time')
plt.xlabel('Amp')
```
